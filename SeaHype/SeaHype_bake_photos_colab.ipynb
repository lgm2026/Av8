# SeaHype — bake real photos in (no installs)

This runs entirely in your browser on Google's free servers. It downloads a real, openly-licensed, watermark-free photo of the **exact species** for each lesson and bakes them **into the app file** — the finished page shows photos from inside itself and never calls another site.

**How to use:** Runtime → **Run all** (or run each cell top to bottom). When asked, upload the `SeaHype-Marine-Biology-Education-FULL-BUNDLE.zip` Claude gave you. ~10 minutes later it auto-downloads the finished HTML.

No Node, no setup — just Python + Pillow, installed for you below.

```python
# 1) Upload the SeaHype bundle .zip and step into it
from google.colab import files
import zipfile, glob, os
print('Upload SeaHype-Marine-Biology-Education-FULL-BUNDLE.zip ...')
up = files.upload()
zname = [k for k in up if k.lower().endswith('.zip')][0]
with zipfile.ZipFile(zname) as z: z.extractall('seahype')
base = os.path.dirname(glob.glob('seahype/**/bake_photos.py', recursive=True)[0])
os.chdir(base)
print('Working in:', os.getcwd())
```

```python
# 2) Install the one dependency
!pip -q install Pillow
```

```python
# 3) Fetch photos + bake them in (~10 min, resumable)
#    This is a FREE site, so the default licenses (incl. CC BY-NC) are used.
#    For a PAID app instead, change LICENSE to: cc0,cc-by,cc-by-sa
LICENSE = 'cc0,cc-by,cc-by-sa,cc-by-nc,cc-by-nc-sa'
!python3 bake_photos.py --license "$LICENSE"
```

```python
# 4) Download the finished, self-contained app (photos baked in)
from google.colab import files
import glob
out = glob.glob('*with photos*.html')[0]
print('Downloading:', out)
files.download(out)
```

**That's it.** Open the downloaded `SeaHype Marine Biology Education (with photos).html` — it's one self-contained file. Photo credits are listed in-app under **Library → Sources → Photo credits**.

Want to spot-check likeness first? The `photos/_credits.json` in the working folder lists the exact species matched for each image; anything in `photos/_missing.json` simply keeps its illustration.
